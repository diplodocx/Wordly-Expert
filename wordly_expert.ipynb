```python
import pandas as pd
```

```python
df = pd.read_csv('./Nouns.csv', sep=';')
df
```

```python
df['word'] = df['word'].str.replace(' ', '')
df['word'] = df['word'].str.replace('-', '')
df['word'] = df['word'].str.replace('ё', 'е')
df['word'] = df['word'].str.replace('c', 'с')
df_w = df['word']
df_w.drop_duplicates(inplace=True)
df_w
```

```python
arr = []
for word in df_w:
    letters = list(word)
    for letter in letters:
        if letter not in arr:
            arr.append(letter)
print(len(arr))
arr = sorted(arr)
for i in arr:
    print(ord(i))
```

```python
df_w.apply(lambda x: len(x)).max()
```

```python
letters = [chr(l) for l in range(1072, 1104)]
dct = {l: [0]*30 for l in letters}
df_res = pd.DataFrame(dct)
df_res
```

```python
for word in df_w:
    letters = list(word)
    for letter in letters:
        length = len(word)
        df_res[letter][length] += 1

df_res
```

```python
dfs = []
letters = [chr(l) for l in range(1072, 1104)]
for i in range(4, 9):
    cur_df = pd.DataFrame({'letter': letters, 'count': list(df_res.iloc[i, :])})
    sum = cur_df['count'].sum()
    cur_df['frequency'] = round(cur_df['count']/sum*100, 2)
    cur_df.sort_values(by='frequency', inplace=True, ascending=False)
    print(cur_df.count())
    dfs.append(cur_df)

df_res_4, df_res_5, df_res_6, df_res_7, df_res_8 = dfs
               
```

```python
for i, df_i in enumerate(dfs):
    print(i+4, df_i['count'].sum())
```
